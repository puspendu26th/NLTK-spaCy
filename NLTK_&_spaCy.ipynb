```python
import nltk
nltk.download('all')
text="The fate of Lehman Brothers, the beleaguered investment bank, hung in the balance on Sunday asFederal Reserve officials and the leaders of major financial institutions continued to gather in emergency meetings trying to complete a plan to rescue the stricken bank. Several possible plans emerged from the talks, held at the Federal Reserve Bank of New York and led by Timothy R. Geithner,the president of the New York Fed, and Treasury Secretary Henry M. Paulson Jr. "



```

B. Perform the following operations upon the given text using NLTK:
1. Tokenization
2. Lower case conversion
3. Stop Words removal
4. Stemming
5. Lemmatization
6. Parse tree or Syntax Tree generation
7. POS Tagging 


```python
#1.Tokenization
for i in nltk.word_tokenize(text):
  print(i)
sentence=nltk.sent_tokenize(text)
sentence
```

```python
#2.Lower case conversion
lower_case=text.lower()
print(lower_case)
```

```python
#3.Stop Words removal
from nltk.corpus import stopwords
word=nltk.word_tokenize(text)
stop_words=set(stopwords.words("english"))
new_array=[]
for i in word:
  if i.casefold() not in word:
    new_array.append(i)
print(new_array)
```

```python
#4.Stemming
from nltk.stem import PorterStemmer
ps=PorterStemmer()
word=nltk.word_tokenize(text)
for i in word:
  print(ps.stem(i))
```

```python
#5.Lemmatization
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
word=nltk.word_tokenize(text)
for i in word:
  print(lemma.lemmatize(i))
 
```

```python
from nltk.corpus.reader import word_tokenize
import matplotlib
#6.Parse tree or Syntax Tree generation 
word=nltk.word_tokenize(text)
tagg=nltk.pos_tag(word)
grammar = """
    NP: {<DT|PP\$>?<JJ>*<NN>}   # chunk determiner/possessive, adjectives and noun
    P : {<IN>}                  # chunk prepositions
    V  : {<V.*>}                # chunk verbs
    PP : {<P> <NP>}             # chunk prepositions followed by NP
    VP : {<V> <NP|PP>*}         # chunk verbs and their arguments
"""
chunk=nltk.RegexpParser( grammar )
 
   
                          
                      
tree=chunk.parse(tagg)
print(tree)





```

```python
#7.POS Tagging
word=nltk.word_tokenize(text)
tagg=nltk.pos_tag(word)
tagg
```

A. Perform the following operations upon the given text using spaCy:
1. Tokenization
2. Parts-of-speech tagging
3. Dependency parsing
4. Named Entity Recognition
5. Lemmatization 


```python
#1.Tokenization
import spacy
nlp = spacy.load('en_core_web_sm')
doc=nlp("The fate of Lehman Brothers, the beleaguered investment bank, hung in the balance on Sunday asFederal Reserve officials and the leaders of major financial institutions continued to gather in emergency meetings trying to complete a plan to rescue the stricken bank. Several possible plans emerged from the talks, held at the Federal Reserve Bank of New York and led by Timothy R. Geithner,the president of the New York Fed, and Treasury Secretary Henry M. Paulson Jr. ")

for i in doc:
  print(i.text)
```

```python
#2.Parts-of-speech tagging
for i in doc:
  print(i.text,i.pos_)
```

```python
#3.Dependency parsing
for i in doc:
  print(i.text,i.dep_)
```

```python
#4.Named Entity Recognition
for i in doc.ents:
  print(i.text,i.start_char,i.end_char,i.label_)
```

```python
#5.Lemmatization
for i in doc:
  print(i.text,i.lemma_)
```
